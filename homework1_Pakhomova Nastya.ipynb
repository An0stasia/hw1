```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib
```

```python
df = pd.read_csv("Pearson.tsv", sep = "\t")
df.head()
```

```python
x = df["Father"]
xx = df[df["Father"] == 67.8]["Son"]
xxx = df[df["Father"] == 68.3]["Son"]
xxxx = df[df["Father"] == 66.0]["Son"]
y = df["Son"]
yy = df[df["Son"] == 68.6]["Father"]
yyy = df[df["Son"] == 67.2]["Father"]
yyyy = df[df["Son"] == 66.3]["Father"]

matplotlib.rcParams['figure.figsize'] = (12, 9)
grid = plt.GridSpec(7, 7, wspace=0.4, hspace=0.9)
general_ax = plt.subplot(grid[0:4, 0:4])
ax_1 = plt.subplot(grid[4, 0:4])
ax_2 =  plt.subplot(grid[5, 0:4])
ax_3 = plt.subplot(grid[6, 0:4])
y_1 = plt.subplot(grid[0:4, 4])
y_2 = plt.subplot(grid[0:4, 5])
y_3 = plt.subplot(grid[0:4, 6])

matplotlib.rcParams['font.family'] = 'serif'

general_ax.scatter(x, y, c = y,  cmap = 'RdBu')
general_ax.set_title('Fathers heights and sons heights', fontsize= 10,  fontweight= 'bold')
ax_1.hist(xx, bins=10, color = "coral")
ax_1.set_title('67.8 inch father', fontsize= 9)
ax_2.hist(xxx, bins=10, color = "coral")
ax_2.set_title('68.3 inch father', fontsize= 9)
ax_3.hist(xxxx, bins=10, color = "coral")
ax_3.set_title('66 inch father', fontsize= 9)
y_1.set_title('68.6 inch son', fontsize= 9)
y_1.hist(yy, bins=10, orientation = "horizontal")
y_2.set_title('67.2 inch son', fontsize= 9)
y_2.hist(yyy, bins=10, orientation = "horizontal")
y_3.set_title('66.3 inch son', fontsize= 9)
y_3.hist(yyyy, bins=10, orientation = "horizontal")
```
